```python
#Diabetes Outcome Analysis - Pandas and Matplotlib/Seaborn
#This project uses a real-world diabetes dataset to explore the distribution of diabetic vs. non-diabetic cases, and analyze feature relationships using pie charts, bar charts, and box plots


#Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Load data
filename = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%205/data/diabetes.csv"
df = pd.read_csv(filename)
print(df)

#Check for missing values
print(df.isnull().sum())

#Outcome Counts as DataFrame
outcome_counts = df['Outcome'].value_counts()
outcome_df = pd.DataFrame(outcome_counts.values,
                          index = ['Not Diabetic', 'Diabetic'],
                          columns = ['Count'])
print(outcome_df)

#Pie chart - Diabetes distribution
labels = 'Not Diabetic', 'Diabetic'
colors = '#66b3ff', '#ff9999'
explode = (0, 0.1)

plt.figure(figsize=(6,6))
plt.pie(outcome_counts, labels=labels, colors=colors, autopct='%1.1f%%', explode=explode, shadow=True, startangle=140)
plt.title('Diabetes Outcome Distribution')
plt.axis('equal')
plt.show()

#Bar Chart - outcome counts
outcome_counts.plot(kind = 'bar', color = colors)
plt.title('Diabetes Outcome count')
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.xticks([0,1], labels, rotation=0)
plt.show()

#Box plot - BMI vs. Outcome
sns.boxplot(x='Outcome', y='BMI',hue = 'Outcome', data = df, palette = colors, legend = False)
plt.title('BMI distribution by Diabetes Outcome')
plt.xlabel('Outcome')
plt.ylabel('BMI')
plt.xticks([0,1], labels)
plt.show()

#Box plot - Glucose vs Outcome
sns.boxplot(x='Outcome', y='Glucose',hue = 'Outcome', data = df, palette = colors, legend = False)
plt.title('Glucose Levels by Diabetes Outcome')
plt.xlabel('Outcome')
plt.ylabel('Glucose')
plt.xticks([0, 1], labels)
plt.show()

"""Insights(
- Approximately 35% of the patients in the dataset are diabetic.
- Diabetic individuals tend to have higher BMI and glucose values on average.
- This analysis reveals clear patterns that could inform early health interventions.)"""
```
